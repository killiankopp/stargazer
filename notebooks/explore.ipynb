````python
import os
import requests
from pprint import pprint
````

````python
USER = 'killiankopp'
REPO = 'stargazer'
````

````python
url = f"https://api.github.com/repos/{USER}/{REPO}/stargazers"
headers = {
    "Accept": "application/vnd.github+json",
    "X-GitHub-Api-Version": "2022-11-28"
}
````

````python
def api_call(url):
    response = requests.get(url, headers=headers)

    return response.json()
````

````python
def smart_lister(datas, field):
    list = []
    for data in datas:
        list.append(data[field])

    return list
````

## List stargazers
- Endpoint : https://api.github.com/repos/OWNER/REPO/stargazers
- Docs : https://docs.github.com/en/rest/activity/starring?apiVersion=2022-11-28#list-stargazers

````python
def get_stargazers(user, repo):
    url = f"https://api.github.com/repos/{user}/{repo}/stargazers"
    datas = api_call(url)

    return datas
````

````python
user = 'killiankopp'
repo = 'stargazer'
datas = get_stargazers(user, repo)
pprint(datas)
````

````python
user = 'PyGithub'
repo = 'PyGithub'
datas = get_stargazers(user, repo)
listing = smart_lister(datas, 'login')
print(listing)
````

##  List repositories starred by a user

````python
username = 'killiankopp'
url = f"https://api.github.com/users/{username}/starred"
datas = api_call(url)
pprint(datas)
````
